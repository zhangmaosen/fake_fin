```python
from ollama import Client
client = Client(
  host='http://100.103.46.96:11434'
)
```

```python
from pathlib import Path
import json

folder_path = Path('data_china_new')

out_folder_path = Path('data_china_new_translated')

for file_path in folder_path.glob('*.json'):
    with open(file_path, 'r', encoding='utf-8') as file:
        data = json.load(file)
        print(f"Processing file: {file_path}")
        # 在这里添加对 JSON 数据的操作
        response = client.chat(model='qwq:32b', messages=[{
            'role': 'system',
            'content': 'You are a translator about finance content. your task is to translate the content into English.',

        },
            {
                'role': 'user',
                'content': data['content'],
            },
            ])
        data['content'] = response['message']['content']
        del data['title']

        with open(out_folder_path / file_path.name, 'w', encoding='utf-8') as out_file:
            json.dump(data, out_file, ensure_ascii=False, indent=4)
```
