```python
from ollama import Client
client = Client(
  host='http://100.103.46.96:11434'
)
```

```python
from pathlib import Path
import json

folder_path = Path('../storage\datasets\JGDY_Crawl_Detail')

out_folder_path = Path('JGDY_Crawl_Detail_EN')

for file_path in folder_path.glob('*.json'):
    with open(file_path, 'r', encoding='utf-8') as file:
        data = json.load(file)
        print(f"Processing file: {file_path}")
        # 在这里添加对 JSON 数据的操作
        response = client.chat(model='qwen2.5:32b-instruct', messages=[{
            'role': 'system',
            'content': 'You are a translator. your must translate the input content into English. directly output the translated content.',

        },
            {
                'role': 'user',
                'content': data['content'],
            },
            ],options={'num_ctx': 20000,})
        data['content'] = f"{response['message']['content']}"
        del data['title']

        with open(out_folder_path / file_path.name, 'w', encoding='utf-8') as out_file:
            json.dump(data, out_file, ensure_ascii=False, indent=4)
```
